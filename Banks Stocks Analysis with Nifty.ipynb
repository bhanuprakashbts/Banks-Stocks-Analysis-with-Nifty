```python
#Import Required Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib

import nsepy
from datetime import date
from nsepy import get_history
import seaborn as sns
```

Requesting stock information using nsepy package

```python
sbin = get_history(symbol='SBIN',
                   start=date(2017,10,15),
                   end=date(2018,10,15))

icici = get_history(symbol='ICICIBANK',
                   start=date(2017,10,15),
                   end=date(2018,10,15))

hdfcbank = get_history(symbol='HDFCBANK',
                   start=date(2017,5,28),
                   end=date(2018,5,28))

nifty = get_history(symbol='NIFTY',index = True,
                   start=date(2017,10,15),
                   end=date(2018,10,15))
```

Calculation of Beta of Stocks

Calculation of Beta of SBIN

```python
# Beta of SBIN
sbin_beta_data = pd.DataFrame({"nifty":nifty.Close,"sbin":sbin.Close})
sbin_beta_data.head()
```

```python
sbin_beta_data_returns = sbin_beta_data.pct_change()
sbin_beta_data_returns.head()
```

```python
#Covariance Calculation
sbin_beta_data_cov = sbin_beta_data_returns.cov()*250
print(sbin_beta_data_cov)
```

```python
sbin_cov_with_market = sbin_beta_data_cov.iloc[0,1]
print(sbin_cov_with_market)
```

```python
market_variance = sbin_beta_data_returns['nifty'].var()*250
print(market_variance)
```

```python
sbin_beta = sbin_cov_with_market/market_variance
print(round(sbin_beta,2))
```

Calculation of Beta of ICICI BANK

```python
icici_beta_data = pd.DataFrame({"nifty":nifty.Close,"icici":icici.Close})
icici_beta_data_returns = icici_beta_data.pct_change()
icici_beta_data_cov = icici_beta_data_returns.cov()*250
icici_cov_with_market = icici_beta_data_cov.iloc[0,1]
print("Covariance of ICICI with Market is "+ str(icici_cov_with_market))
market_variance = icici_beta_data_returns['nifty'].var()*250
print("Market Variance is "+ str(market_variance))
icici_beta = icici_cov_with_market/market_variance
print("Beta of ICICI is " + str(round(icici_beta,2)))
```

Calculation of Beta of HDFC BANK

```python
hdfcbank_beta_data = pd.DataFrame({"nifty":nifty.Close,"hdfcbank":hdfcbank.Close})
hdfcbank_beta_data_returns = hdfcbank_beta_data.pct_change()
hdfcbank_beta_data_cov = hdfcbank_beta_data_returns.cov()*250
hdfcbank_cov_with_market = hdfcbank_beta_data_cov.iloc[0,1]
print("Covariance of HDFC BANK with Market is "+ str(hdfcbank_cov_with_market))
market_variance = hdfcbank_beta_data_returns['nifty'].var()*250
print("Market Variance is "+ str(market_variance))
hdfcbank_beta = hdfcbank_cov_with_market/market_variance
print("Beta of HDFC BANK is " + str(round(hdfcbank_beta,2)))
```
